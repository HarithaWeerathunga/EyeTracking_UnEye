```python
import pandas as pd
import csv
import math 
import numpy as np

import uneye



```

```python
pd.options.display.max_columns = None

## get the data of subjects 2,6,8,14,21,22,28,34

data_file = pd.read_csv('train.csv',header=None, on_bad_lines='skip', engine='python')


filtered_dataframe = data_file.loc[(data_file[0] == 's2') | ( data_file[0] == 's6') | ( data_file[0] == 's8') | ( data_file[0] == 's14') 
| ( data_file[0] == 's21')| ( data_file[0] == 's22')| ( data_file[0] == 's28') | ( data_file[0] == 's34')]
```

```python
subjects = ['s2', 's6', 's8', 's14', 's21', 's22', 's28', 's34']
```

```python
def create_subjects_array():
    result = []
    for s in subjects:
        subject_data = filtered_dataframe.loc[filtered_dataframe[0] == s]
        subject_dict = {}
        data_false = subject_data.loc[subject_data[1] == False]
        data_true = subject_data.loc[subject_data[1] == True]
        subject_dict['subject_name'] = s
        subject_dict['known_true'] = []
        subject_dict['known_false'] = []
        create_coordinates(data_false, subject_dict['known_false'])
        create_coordinates(data_true, subject_dict['known_true'])

        result.append(subject_dict)
    return result 
    
def create_coordinates(data, subject_dict):
  for index, row in data.iterrows():
      row = row[2:]
      subject_dict.extend([*zip(row[::2], row[1::2])])
```

```python
subjects = create_subjects_array()
```

```python
import numpy as np

import matplotlib.pyplot as plt

##################################
########### Parameters ###########
sampfreq = 1000 #Hz
weights_name = 'weights_synthetic'
min_sacc_dur = 6 # in ms
# min_sacc_dist = 10 #in ms
min_sacc_dist = 5 #in ms

# Prediction
model = uneye.DNN(weights_name=weights_name,
                        sampfreq=sampfreq,
                        min_sacc_dur=min_sacc_dur,
                        min_sacc_dist=min_sacc_dist)
```

```python


def getEvents(subjects):    

    for each in subjects:
        X_known = []
        Y_known = []

        X_unknown = []
        Y_unknown = []

        events_true = each['known_true']
        events_false = each['known_false']
        subject_name = each['subject_name']
        for value in events_true:
            if math.isnan(value[0]) | math.isnan(value[1]):
                continue
            else:
                X_known.append(value[0])
                Y_known.append(value[1])

        print(subject_name)

        Prediction_Known,Probability_Known = model.predict(np.array(X_known),np.array(Y_known))
        
        # print(type(Prediction_Known))
        unique, counts = np.unique(Prediction_Known, return_counts=True)
        output_known = dict(zip(unique, counts))
        # print(output_known)
        saccade_count_known = output_known[0.0]
        fixation_count_known = output_known[1.0]

        

        for value in events_false:
            if math.isnan(value[0]) | math.isnan(value[1]):
                continue
            else:
                X_unknown.append(value[0])
                Y_unknown.append(value[1])


        Prediction_Unkown,Probability_Unknown = model.predict(np.array(X_unknown),np.array(Y_unknown))
        unique, counts = np.unique(Prediction_Unkown, return_counts=True)
        output_unknown = dict(zip(unique, counts))

        saccade_count_unknown = output_unknown[0.0]
        fixation_count_unknown = output_unknown[1.0]

        print('Subject' , str(subject_name))
        # print(Prediction_Known)
        print("Saccade Count Known " + str(saccade_count_known))
        print("Fixation Count Known " + str(fixation_count_known))

        # print(Prediction_Unkown)
        print("Saccade Count Unknown " + str(saccade_count_unknown))
        print("Fixation Count Unknown " + str(fixation_count_unknown))

        print("*************************************************")


    
        






    
```
